```python
import matplotlib.pyplot as plt

years = list(range(2020, 2026))  # 2019, 2020, 2021, 2022, 2023, 2024
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

for ax, year in zip(axes, years):
    # fetch and convert to pandas
    df = get_binance_trades_hourly_summary(year).to_pandas()

    ax.hist(df['quantity_per_trade'], bins=1000)
    ax.set_title(str(year))

    ax.set_xlim(0, 0.2)
    ax.set_ylim(0, 80)
    
    ax.set_xlabel('Quantity per trade')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = list(range(2020, 2026))  # 2019, 2020, 2021, 2022, 2023, 2024
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

for ax, year in zip(axes, years):
    # fetch and convert to pandas
    df = get_binance_trades_hourly_summary(year).to_pandas()

    ax.hist(df['total_trades'], bins=5)
    ax.set_title(str(year))

    ax.set_xlim(0, 1500000)
    ax.set_ylim(0, 10000)
    
    ax.set_xlabel('total_trades')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
