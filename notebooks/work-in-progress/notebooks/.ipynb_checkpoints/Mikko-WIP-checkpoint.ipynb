```python
import pyarrow.csv as csv
import pyarrow as pa
import pyarrow.dataset as ds
from pathlib import Path
import tqdm

folder = Path("/opt/trades-warehouse-backup/data/")

schema = pa.schema([
    ("trade_id", pa.int64()),
    ("price", pa.float32()),
    ("quantity", pa.float32()),
    ("quote_quantity", pa.float32()),
    ("timestamp", pa.int64()),
    ("is_buyer_maker", pa.string()),
    ("is_best_match", pa.string()),
    
])

tables = []
for file in tqdm.tqdm(folder.glob(f"BTCUSDT-trades-2024-{}-*.csv")):
    table = csv.read_csv(file,
                         read_options=csv.ReadOptions(column_names=schema.names),
                         convert_options=csv.ConvertOptions(column_types=schema))
    tables.append(table)

full_table = pa.concat_tables(tables)
df = pl.from_arrow(full_table)
```

## Multi-frame parameter sweep
Takes as its input framing functions and their parameters. Then performs parameter sweep for each, where the resulting data from each frame is then used to perform rolling correlation between high price and RSI. Finally, this is plotted individually by frame.

```python
import numpy as np

from rsi_rolling_corr_loop import rsi_rolling_corr_loop
from frame_by_zigzag_threshold import frame_by_zigzag_threshold # 1.0001 - 2.0000
from frame_by_time_interval import frame_by_time_interval # 1s, 2s, 3s, 4s, 5s,
from frame_by_price_range import frame_by_price_range # 0.001 - 0.050 

params = {
    'frame_by_zigzag_threshold': (1.0001, 2.0000, 0.0001), # min, max, step
    'frame_by_time_interval': (1, 10, 1), 
    'frame_by_price_range': (0.0001, 0.05, 0.0001)}

sample = 100

# 

for framing_function in [frame_by_price_range, frame_by_zigzag_threshold, frame_by_time_interval]:

    title = f'{framing_function.__name__}: pl.rolling_cor of `price` with `RSI` \n (Sample: n={sample} parameter iterations.)'

    frame_params = np.random.choice(np.arange(*params[framing_function.__name__]), size=sample)
    
    results = rsi_rolling_corr_loop(df=df_datetime,
                                    framing_function=framing_function,
                                    params=frame_params,
                                    n=sample,
                                    plot_title=title)
```
